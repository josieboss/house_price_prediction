```python
df['date'] = pd.to_datetime(df['date'])
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df = df.drop('month', axis=1)
df = df.drop('year', axis=1)
df = df.drop('day', axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.insert(2, 'day', df['date'].dt.day)
df.insert(3, 'month', df['date'].dt.month)
df.insert(4, 'year', df['date'].dt.year)
```

```python
df.head()
```

```python
df['condition'].unique()
```

```python
def replace_val(x):
    curr = {1:'Bad', 2: 'Not_Good', 3: 'Good', 4: 'Very_Good', 5: 'Excellent'}
    return curr[x]
```

```python
replace_val(5)
```

```python
df['condition'] = df['condition'].apply(replace_val)
```

```python
val = []
for key in df['condition'].value_counts().keys():
    val.append(df['condition'].value_counts()[key])
```

```python
val
```

```python
import matplotlib.pyplot as plt
```

```python
courses = list(df['condition'].value_counts().keys())
values = val

fig = plt.figure(figsize = (10,5))
plt.bar(courses, values, color = 'purple', width = 0.4)

plt.xlabel('House conditions')
plt.ylabel('count')
plt.title('House conditions distributions')
plt.show
```

```python
df['date'] = pd.to_datetime(df['date'])
df.insert(2, 'day', df['date'].dt.day)
df.insert(3, 'month', df['date'].dt.month)
df.insert(4, 'year', df['date'].dt.year)
```

```python
def replace_val(x):
    curr = {1:'Bad', 2: 'Not_Good', 3: 'Good', 4: 'Very_Good', 5: 'Excellent'}
    return curr[x]
```

```python
replace_val(5)
```

```python
df['condition'] = df['condition'].apply(replace_val)
```
